```python
import pandas as pd
import numpy as np
from math import sqrt
from scipy.stats.stats import pearsonr   
from scipy.stats import norm

import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
pio.renderers.default = "notebook_connected"
import matplotlib.pyplot as plt

import csv
import os
import time
import websockets as ws
from datetime import datetime

#using https://github.com/twopirllc/pandas-ta
import pandas_ta as ta

# https://python-binance.readthedocs.io/en/latest/overview.html
try: 
    from binance.client import Client
except:
    time.sleep(1)
    from binance.client import Client
```

```python
api_key = 0
api_secret = 0
client = Client(api_key, api_secret)

# all markets
info = client.get_all_tickers()

def load_symbol(symbol, timeframe, limit):
    candles = client.get_klines(symbol=symbol, interval=timeframe, limit=limit)
    df = pd.DataFrame(candles)
    df = df.rename(columns={
        0: 'open_time', 
        1: 'open',
        2: 'high',
        3: 'low',
        4: 'close',
        5: 'volume',
        6: 'close_time',
        7: 'quote_asset_volume',
        8: 'number_of_trades',
        9: 'taker_buy_base_asset_volume',
        10: 'taker_buy_quote_asset_volume',
        11: 'ignore',
    })
    df = df.drop(columns=['ignore'])
    df = df.astype(float)
    df['datetime'] = pd.to_datetime(df['open_time'], unit='ms')
    df = df.sort_values(by=['datetime'])
    
    return df

def load_all_symbols(symbol_list, timeframe, limit):
    all_symbols_dict = {}
    
    for symbol in symbol_list:
        df = load_symbol(symbol, timeframe, limit)
        all_symbols_dict[symbol] = df
        
    return all_symbols_dict
```

```python
symbol_list_usdt = ['BTCUSDT', 'ETHUSDT']
all_symbols_1day = load_all_symbols(symbol_list_usdt, Client.KLINE_INTERVAL_1DAY, 1096) # 2 years of data + 1

btc_df = all_symbols_1day['BTCUSDT']
data= btc_df['close']
```

# Monte Carlo

```python
# Based on https://datascienceplus.com/how-to-apply-monte-carlo-simulation-to-forecast-stock-prices-using-python/

log_returns = np.log(1 + data.pct_change())
print(log_returns.tail())
data.plot(figsize=(10, 6));
```

```python
log_returns.plot(figsize = (10, 6))
```

```python
u = log_returns.mean()
var = log_returns.var()
drift = u - (0.5 * var)
stdev = log_returns.std()


print(u)
print(var)
print(drift)
print(stdev)
```

```python
# run simulation of daily returns
t_intervals = 365 # intervals, thus how much data points in time
iterations = 10000 # how much simulations I want to do 
daily_returns = np.exp(drift + stdev * norm.ppf(np.random.rand(t_intervals, iterations)))

# get last price and place as first data point in the price list
S0 = data.iloc[-1]
price_list = np.zeros_like(daily_returns)
price_list[0] = S0

# fill price list
for t in range(1, t_intervals):
    price_list[t] = price_list[t-1] * daily_returns[t]
```

```python
# take last prices from the price_list
last_prices = []
for i in price_list:
    last_prices.append(price_list[-1])
    
```

```python
max_price = price_list[-1].max()
min_price = price_list[-1].min()
average_price = np.mean(price_list[-1])
stdev_price = np.std(price_list[-1])


print("Maximum price:", round(max_price, 2))
print("Minimum price:", round(min_price, 2))
print("Average price:", round(average_price, 2))
print("Stdev price:", round(stdev_price, 2))
```

```python
# check what percentage is below a number
orig_len = float(len(price_list[-1])) # float used to avoid integer division
interesting_len = len([i for i in price_list[-1] if i < 50000])
percent = interesting_len/orig_len
print(percent)
```
